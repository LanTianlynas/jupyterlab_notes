# zip()函数
zip() 函数用于将可迭代的对象作为参数，将对象中对应的元素打包成一个个元组，然后返回由这些元组组成的对象，这样做的好处是节约了不少的内存。

我们可以使用 list() 转换来输出列表。

如果各个迭代器的元素个数不一致，则返回列表长度与最短的对象相同，利用 * 号操作符，可以将元组解压为列表。

```python
a = ["a1", "a2", "a3"]
b = ["b1", "b2", "b3"]
c = ["c1", "c2", "c3", "c4", "c5"]
```

```python
zipped_ab = zip(a, b)
```

```python
zipped_ac = zip(a, c)
```

```python
zipped_abc = zip(a, b, c)
```

```python
print(zip(a, b))
```

```python
list(zip(a,b))
```

```python
print(*zip(a, b))
```

```python
[zip for zip in [*zip(a,b)]]
```

```python
print(zip(*zip(a, b)))
```

```python
list(zip(*zip(a, b)))
```

```python
list(zip(zip(a, b)))
```

## 无限Iterators

### count()

##### 1. count(0)

```python
from itertools import count
counter = count(0)
```

```python
for i in counter:
    print(i)
    if i > 9:
        break
```

---

##### 2. count(0,5)

### cycle()

```python
from itertools import cycle
List = ["Lantian", "Sunyang", "Lanxintian"]
cycler = cycle(List)
```

```python
for i, person in enumerate(cycler):
    print(f"{i}:{person}")
    if i == 10:
        break
    
```

### repeat()

```python
from itertools import repeat
str = "I love you"
repeater = repeat(str, 10)
```

```python
for str in repeater:
    print(str)
```

## other01

#### accumulate()

##### accumulate([])

```python
from itertools import accumulate
numbers = range(1, 11)
accumulation = accumulate(numbers)
print(list(accumulation))
```

##### accumulate([], operator.mul)

```python
from itertools import accumulate
from operator import mul
```

```python
numbers = range(1, 11)
accumulation_mul = accumulate(numbers, func=mul)
print(list(accumulation_mul))
```

#### chain()

```python
from itertools import chain
a = range(100, 106)
b = ["a", "b", "c", "d"]
c = ["lantian", "sunyang"]
```

```python
long = chain(a, b, c, c, b, a)
print(list(long))
```
